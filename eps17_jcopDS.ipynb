```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = np.sqrt(x)
y1 = np.sin(x)
y2 = np.cos(x)
```

**Inisialisasi Figure dan Axes**

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y1)
ax.plot(x, y2)
```

**SHORTCUT**

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y1, 'r-', linewidth=0.5) #ketebalan garis
ax.plot(x, y2, 'bo', markersize=1) #ukuran simbol
ax.set_xlim(0,10)
ax.set_ylim(-2,2)
ax.set_title("sin dan cos", fontsize=14)
ax.set_xlabel('x', fontsize=10)
ax.set_ylabel('y', fontsize=10)
ax.axis('equal') # memaksa agar menjadi equal, skalanya menjadi sama
plt.title("sin dan cos")
```

```python
from matplotlib.patches import Shadow
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.plot(x, y1, 'r-', linewidth=0.5, label='Sin')
ax.plot(x, y2, 'bo', markersize=3, label='Cos')
ax.set(xlim=(0, 10), ylim=(-2,2), title="sin dan cos", xlabel="x", ylabel="y")
ax.axis('equal')
ax.legend(loc='lower center', ncol=2, shadow=True, borderpad=1)
```

**SUBPLOT**

**MATLAB STYLE**

```python
plt.figure(figsize=(15,10))
plt.subplot(231) #231 ==> 2 baris 3 kolom bagian pertama
plt.hist(numbers, bins=80, density=1);
plt.subplot(233) #233 ==> 2 baris 3 kolom bagian ketiga
plt.hist(numbers, bins=80, density=1);
plt.subplot(235)
plt.scatter(X, Y, c='r', s=10)
plt.axis('equal')
plt.subplot(236)
plt.hist(numbers, bins=80, density=1);
```

**MATPLOTLIB STYLE**

```python
fig, ax = plt.subplots(2, 3, figsize=(15,10))
ax[0,0].hist(numbers, bins=80, density=1);
ax[0,1].axis('off') # cara menghilangkan kolom
ax[1,0].axis('off') # cara menghilangkan kolom
ax[0,2].hist(numbers, bins=80, density=1);
ax[1,1].scatter(X, Y, c='r', s=10)
ax[1,1].axis('equal');
ax[1,2].hist(numbers, bins=80, density=1);
```

```python
fig, ax = plt.subplots(2, 3, figsize=(15,10), sharex=True, sharey=True)
#sharex/y berfungsi untuk membuat subplot dalam satu figure berbagi sumbu x atau sumbu y
#semua subplot akan menggunakan skala sumbu yang samaa, sehingga hanya ada satu acuan
#sumbu x atau y yang dipakai, maka dari itu equal tidak berlaku jika menggunakan sharex/y
ax[0,0].hist(numbers, bins=80, density=1);
# ax[0,1].axis('off') # cara menghilangkan kolom
# ax[1,0].axis('off') # cara menghilangkan kolom
ax[0,2].hist(numbers, bins=80, density=1);
ax[1,1].scatter(X, Y, c='r', s=10)
ax[1,2].hist(numbers, bins=80, density=1);
```

**3D PLOT**

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
```

```python
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2 + 10)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, antialiased=False, cmap='winter')
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.view_init(elev=15, azim=25)
# elev atas bawah, azimut kiri kanan
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='winter')
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.view_init(elev=-15, azim=25)
# elev atas bawah, azimut kiri kanan
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='winter')
ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.view_init(elev=-15, azim=25)
# elev atas bawah, azimut kiri kanan
```
